# Feature Engineering

This section focues on transforming the time data saved on minutes to an actual format and create proper columns to save the correct format data

```python
flights['Elapsed_Hours'] = (flights['CRS_Elapsed_Time'] // 60).astype(int)
flights['Elapsed_Minutes'] = (flights['CRS_Elapsed_Time'] % 60).astype(int)
flights['CRS_Elapsed_Time_HHMM'] = flights['Elapsed_Hours'].astype(str).str.zfill(2) + ':' + flights['Elapsed_Minutes'].astype(str).str.zfill(2)
flights['CRS_Elapsed_Time_HHMM']
```

```python
flights=flights.drop(columns=['Elapsed_Hours', 'Elapsed_Minutes'])
```

```python
#transforming time data to a correct format
flights['Hours'] = (flights['Actual_Elasped_Time'] // 60).astype(int)
flights['Minutes'] = (flights['Actual_Elasped_Time'] % 60).astype(int)
flights['Actual_Elasped_Time_HHMM'] = flights['Hours'].astype(str).str.zfill(2) + ':' + flights['Minutes'].astype(str).str.zfill(2)
flights['Actual_Elasped_Time_HHMM']
```

```python
flights=flights.drop(columns=['Hours', 'Minutes'])
```

```python
flights=flights.drop(columns=['CRS_Elapsed_Time', 'Actual_Elasped_Time'])
```

```python
flights.head(10)
```

```python
#checking if the Arr_Delay column is in correct or contains wrong information
flights["Dep_Delay"]
```

```python
#taking all the negative data of the arr_delay to a correct column containing flights that arrived earlier than expected
flights["Arrived_Late_MM"] = flights["Arr_Delay"].apply(lambda x: x if x > 0 else 0)
```

```python
flights["Arrived_Early_MM"] = flights["Arr_Delay"].apply(lambda x: x if x < 0 else 0)
```

```python
flights["Arrived_Early_MM"]=flights["Arrived_Early_MM"].abs()
```

```python
flights=flights.drop(columns=["Arr_Delay"])
```

```python
flights["Early_Dep_M"] = flights["Dep_Delay"].apply(lambda x: x if x < 0 else 0)
```

```python
flights["Early_Dep_M"]=flights["Early_Dep_M"].abs()
```

```python
flights["Dep_Delay"] = flights["Dep_Delay"].apply(lambda x: x if x > 0 else 0) 
```

```python
flights.head(5)
```

```python
#giving the distance data to a correct format
flights["Distance"].value_counts()
```

```python
#giving the distance data to a correct format & saving it in a new column
flights['Distance_Miles'] = flights['Distance'].str.replace('miles', '', regex=False)
flights['Distance_Miles'] = flights['Distance_Miles'].str.strip()
```

```python
flights['Distance_Miles'] = flights['Distance_Miles'].astype(int)
```

```python
flights=flights.drop(columns=["Distance"])
```
